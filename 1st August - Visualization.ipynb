```python
# o to 10 into 100 equal parts = 0.1
import numpy as np
x = np.linspace(0,10,100)
print(x)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
plt.plot(x, np.sin(x - 0), color='blue')
plt.plot(x, np.sin(x - 1), color='g')
plt.plot(x, np.sin(x - 2), color='0.75')
plt.plot(x, np.sin(x - 3), color='#FFDD44')
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3))
plt.plot(x, np.sin(x - 5), color='chartreuse');
plt.show()
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')
```

```python
plt.plot(x, x + 4, linestyle='solid')
plt.plot(x, x + 5, linestyle='dashed')
plt.plot(x, x + 6, linestyle='dashdot')
plt.plot(x, x + 7, linestyle='dotted')
```

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed
plt.plot(x, x + 2, '-.k') # dashdot
plt.plot(x, x + 3, ':r') # dotted red
```

```python
# Adjusting the plot
plt.plot(x, np.sin(x))
# plt.axhLine(y = 0, Linestyle = '--')
plt.xlim(0,11)
plt.ylim(-1.5,1.5);
```

```python
# automatically tightens the bounds ( Compress)
plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
# An equal aspect ration
plt.plot(x, np.sin(x))
plt.axis('equal')
```

```python
# GG Plot
from matplotlib import style
style.use('ggplot')

a1 = [5,8,10]
b1 = [12,16,6]

a2 = [6,9,11]
b2 = [6,15,7]

plt.plot(a1,b1,'r',label ='line 1',linewidth=4)
plt.plot(a2,b2,'b',label ='line 2',linewidth=4)

plt.title('Epic Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.legend()
plt.grid(color = 'k')
plt.show()
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.title("A sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()
```

# Bar Graph


```python
# Bar Plot shows the categorical values
plt.bar([1,3,5,7,9],[5,2,7,8,2], label ='Instance1')
plt.bar([2,4,6,8,10],[8,6,2,5,6], label ='Instance2', color = 'b')

plt.legend()

plt.xlabel('Bar number')
plt.ylabel('Bar height')

plt.title("Bar plot")

plt.show()
```

```python
# Bar Plot shows the categorical values
plt.bar([1,3,5,7,9],[5,2,7,8,2], label ='Instance1', color = 'y')
plt.bar([2,4,6,8,10],[8,6,2,5,6], label ='Instance2', color = 'b')

plt.legend()

plt.xlabel('Bar number')
plt.ylabel('Frequency')

plt.title("Bar plot")

plt.show()
```

```python
# Scatter plot
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]
plt.scatter(x,y, label='cat',color='k',s=130, marker="*")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Scatter Plot")

plt.legend()
plt.show()
```

```python
# importing seaborn library

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn import datasets
iris = datasets.load_iris()

import pandas as pd
iris_df = pd.DataFrame(iris.data, columns = ['sepal_length','sepal_width','petal_length','petal_width'])
iris_df['target']=pd.Series(iris.target)
```

```python
print(iris_df)
```

```python
sns.scatterplot(x = 'petal_length', y = 'petal_width', data = iris_df)
```

```python
iris_df.plot.scatter(x = 'sepal_length', y = 'petal_length',  c='sepal_width')
```

```python
new_df = iris_df.drop(['target'],axis=1).iloc[0]
new_df.plot.pie(figsize = (5,5))
```

```python
# pairwise scatter plot
# Disadvantage
## cannot visualize in highr dimensional part 3-d , 4-d
#Only visvible to 2d patterns

sns.pairplot(iris_df, hue="target");
plt.show()
```
